```python
def to_int(a):
    a = int(a)
    return a
```

```python
def scaleto100(a):
    """
    Scales the occupancy to a 100% for all the observations which might have occupancy more than 100 due to multiple reasons.
    """
    if a > 1:
        a = 1
    return a
```

```python
def convert_to_string(a):
    """
    Converts the datatype of an observation to 'string' for a particular required computation
    """
    a = str(a)
    return a
```

```python
def assign_part_of_day(x):
    """
    Assigns part of day which is required to scale the transaction occupancy 
    by a scaling factor to resonate more closely to the ground truth
    """
    if np.logical_and(x >= 8, x <= 11):
        return "Morning"
    elif np.logical_and(x >= 12, x <= 16):
        return "Afternoon"
    else:
        return "Evening"
```

```python
def scaling_factor(a):
    """
    Assigns scaling factor to each part of the day - computed in a different notebook
    """
    if a=='Morning':
        return 1.159806
    elif a=='Afternoon':
        return 1.115792
    else:
        return 1.089160
```

```python
def cyclic_features( df ):
    """
    Map each cyclical variable onto a circle such that the lowest value for that variable appears right next to the largest value.
    31 Dec is close to 1 Jan and not opposite (for our analysis)
    We compute the x- and y- component of that point using sin and cos trigonometric functions
    Returns a df with all time features converted to cyclic.
    """
    df['hour_sin'] = np.sin(df.hour_transaction*(2.*np.pi/24))
    df['hour_cos'] = np.cos(df.hour_transaction*(2.*np.pi/24))
    df['month_sin'] = np.sin((df.mm_transaction-1)*(2.*np.pi/12))
    df['month_cos'] = np.cos((df.mm_transaction-1)*(2.*np.pi/12))
    df['day_sin'] = np.sin((df.dd_transaction-1)*(2.*np.pi/30))
    df['day_cos'] = np.cos((df.dd_transaction-1)*(2.*np.pi/30))
    df['dow_sin'] = np.sin((df.dow_transaction-1)*(2.*np.pi/7))
    df['dow_cos'] = np.cos((df.dow_transaction-1)*(2.*np.pi/7))
    df['week_sin'] = np.sin((df.week_transaction-1)*(2.*np.pi/52))
    df['week_cos'] = np.cos((df.week_transaction-1)*(2.*np.pi/52))
    return df
```

```python
def data_preprocessing(df, scaled_y=False, add_noise=True, drop_street=True, drop_loc=True, drop_survey=False,
                       time_cyclic_features=True, datetime_convert_cat=False, fillna_event=True,
                       other_convert_cat=True, fillna_traffic=True, drop_time=True, one_hot_encode=True, segkey_features=True):
    """
    Scales target class - Occupied spots/Total Spots - Adds a new variable called 'target'
    Defines a variable called 'time_stamp' and converts it to datetime (combination of date and time of transaction)
    Adds random gaussian noise to the count variables to avoid them to be mistaken as categorical by the model
    Imputes missing values
    Adds features related to segkey
    Returns the new dataframe with new columns and dropped old values
    """

    # define target class
    y = df['occupied_spots'] / df['total_spots']

    # Scale values to 100% for records having more than 100 occupancy.
    y = y.apply(scaleto100)

    df['target'] = y
    print('Target class added.')

    # define two new variables to club date and time into a single variable
    df['date_string'] = df['date_of_transaction'].apply(convert_to_string)
    df['time_string'] = df['time_of_transaction'].apply(convert_to_string)
    df["time_stamp"] = df["date_string"] + " " + df["time_string"]
    df['time_stamp'] = pd.to_datetime(df['time_stamp'])
    print('Time Stamp field added.')

    df['parkingtimelimitcategory'] = df.parkingtimelimitcategory.apply(to_int)

    # scaling factor
    df['part_of_day'] = df['hour_transaction'].apply(assign_part_of_day)
    df['scaling_factor'] = df['part_of_day'].apply(scaling_factor)

    # Add noise to count vars
    if (add_noise):
        mu, sigma = 0, 0.1
        # creating a noise with the same length as the dataset
        noise = np.random.normal(mu, sigma, df.shape[0])
        df['noise'] = noise

        # adding the random noise to the counts to make them continuous
        list_count_var = ['business_license_count',
                          'parking_lots_garages_count', 'banks_count', 'bars_restaurants_count',
                          'event_venues_count', 'grocery_market_count', 'gyms_count',
                          'hotels_count', 'medical_offices_count', 'museums_count',
                          'salons_count', 'shopping_count']

        for count in list_count_var:
            df[count] = df[count] + df['noise']
        print('Noise added to count variables.')

    if(drop_street):
        df = df.drop(['main_lane', 'between_one', 'between_second'], axis=1)
        print('Street variables dropped.')

    if(datetime_convert_cat):
        time_cat_vars = ['mm_transaction', 'yyyy_transaction', 'dd_transaction', 'dow_transaction',
                         'doy_transaction', 'week_transaction', 'hour_transaction', 'min_transaction', ]
        for i in time_cat_vars:
            df[i] = df[i].astype('category')
        print('Converted all datetime variables to category dtype')

    if(fillna_traffic):
        df.avg_traffic_density = df.avg_traffic_density.fillna(
            df.avg_traffic_density.mean())

    if (fillna_event):
        df.event = df.event.fillna(0)
        df['event_cat'] = np.where(df['event'] == 0, 0, 1)
        df = df.drop('event', axis=1)

    if(other_convert_cat):
        other_cat_car = ['parkingcategory', 'parkingtimelimitcategory',
                         'main_city_street', 'event_cat', 'part_of_day', 'sideofstreet']
        for i in other_cat_car:
            df[i] = df[i].astype('category')
        print('Converted all other variables to category dtype')

    if(drop_loc):
        df = df.drop(['latitude', 'longitude', 'geog'], axis=1)
        print('Location variables dropped')

    if(drop_survey):
        df = df.drop(
            ['total_spots_survey', 'occupied_spots_survey', 'time_of_survey'], axis=1)
        print('Survey variables dropped')

    if (time_cyclic_features):
        df = cyclic_features(df)
        print('Converted time features to cyclic')

    if(drop_time):
        df = df.drop('time_stamp', axis=1)

    if (one_hot_encode):
        df = pd.get_dummies(df, columns=other_cat_car)
        print('Completed one hot encoding cat features.')

    if (segkey_features):
        df['near_waterfront'] = df['segkey'].apply(near_waterfront)
        df['near_downtown'] = df['segkey'].apply(near_downtown)
        df['arterial'] = df['segkey'].apply(arterial)
        df = df.drop('segkey', axis=1)  # dropped segkey
        print('Segkey features added.')

    print('Finishing up..')

    # drop columns that are redundant/not required for analysis
    df = df.drop(['time_of_transaction', 'date_of_transaction', 'date_string', 'time_string', 'occupied_spots', 'total_spots',
                  'paidparkingsubarea', 'paidparkingarea', 'weather_at_hour', 'noise', 'empty_spots'], axis=1)

    # scaled version
    df_scaled_y = df.copy()
    df_scaled_y['target'] = df_scaled_y['target'] * \
        df_scaled_y['scaling_factor']
    df = df.drop('scaling_factor', axis=1)
    df_scaled_y = df_scaled_y.drop('scaling_factor', axis=1)

    pd.set_option('display.max_columns', None)

    if (scaled_y):
        return df_scaled_y
    else:
        return df
```

#### Lagged Features

```python
def lagged(X, prev_30=True, prev_hour=True, prev_day=True):
    """
    Takes preprocessed data with date field and returns dataframe with features 
    for lagged parking occupancy for previous half hour, previous hour, and 
    previous day. Drops variables that cannot be lagged accurately (8am for each 
    day for same day lags and first record of the dataset - 1/3/2017- for previous
    day. 8am is always dropped because of data quality issues.).
    """
    # Data must be ordered to find correct previous value
    X = X.sort_values(['source_element_key', 'yyyy_transaction', 'mm_transaction',
                       'dd_transaction', 'hour_transaction', 'min_transaction'])
    if (prev_30):
        X['previous_30m_occ'] = X.target.shift(periods=1, fill_value=0)

    if (prev_hour):
        X['previous_1h_occ'] = X.target.shift(periods=2, fill_value=0)

    # Remove 8:00 AM for everyday - Will have records from 7PM last day when shifted
    X = X[~np.logical_and(X.hour_transaction == 8, X.min_transaction == 0)]

    if (prev_day):
        X = X.sort_values(by=['source_element_key', 'hour_transaction',
                              'min_transaction', 'yyyy_transaction', 'mm_transaction', 'dd_transaction'])
        X['previous_day_occ'] = X.target.shift(periods=1, fill_value=0)

        # Remove 1/3/2017, first day of dataset
        X['date'] = X['time_stamp'].dt.date
        X = X[pd.to_datetime(X.date) != pd.to_datetime('2017-01-03')]
        # Drop date field since it's not needed for modeling
        X = X.drop('date', axis=1)

    # Drop time stamp and source element key since they should not be included
    # in modeling
    X = X.drop(['time_stamp', 'source_element_key'], axis=1)

    return X
```

```python
def test_train_split_custom (df,month=5,year=2018):
    """
    Splits the dataframe into test and train.
    Returns - X_train, X_test, y_train, y_test
    """
    df['train'] = np.where(np.logical_and(df['mm_transaction'] >= month, df['yyyy_transaction'] == year), 0, 1)
    train = df[df['train'] == 1]
    test = df[df['train'] == 0]
    df = df.drop('train',axis=1)
    
    y_train = train['target']
    X_train = train.drop(['target','train'],axis=1)
    
    y_test = test['target']
    X_test = test.drop(['target','train'],axis=1)
    
    return X_train, X_test, y_train, y_test
```

```python
def linear_reg(X, drop_feature=False, drop: list = None):
    """
    Performs sckit-learn's implementation of Linear Regression
    Internally, splits the data into test and train and fits a linear regression model on the training set.
    Prints MSE and r2 value after evaluating model on the test set and training set 
    and returns the r2 for storing in meterics dictionary
    """

    X_train, X_test, y_train, y_test = test_train_split_custom(X)

    if(drop_feature):
        X_train = X_train.drop(drop, axis=1)
        X_test = X_test.drop(drop, axis=1)
        
    print(X_train.columns)

    clf = LinearRegression()

    clf.fit(X_train, y_train)

    yhat = clf.predict(X_test)
    yhat_train = clf.predict(X_train)

    print('Training Set Performance')
    print("MSE: ", mean_squared_error(y_train, yhat_train))
    print("r2: ", r2_score(y_train, yhat_train))
    
    print('Validation Set Performance')
    print("MSE: ", mean_squared_error(y_test, yhat))
    print("r2: ", r2_score(y_test, yhat))
    
    rms = sqrt(mean_squared_error(y_test, yhat))
    print("rmse: ",rms)

    return r2_score(y_test, yhat), rms, clf
```

```python
def rmse(x,y): return math.sqrt(((x-y)**2).mean())

def random_forest(X, trees=10,njobs=-1, show_plot=True, top_features=5, leaves=1,oobscore=False, min_ss=2):
    """
    Random Forest.
    Inputs: df, hyperparameters (currently only - number of trees)
    Returns: A graph of feature importances, r2_score, topNfeatures
    """
    X_train, X_test, y_train, y_test = test_train_split_custom(X)

    clf = RandomForestRegressor(n_estimators=trees,n_jobs=njobs,min_samples_leaf=leaves, max_features="sqrt", oob_score=oobscore,min_samples_split=min_ss)
    clf.fit(X_train, y_train)
    yhat = clf.predict(X_test)
    yhat_train = clf.predict(X_train)

    print("r_2 score (Training Set): ", r2_score(y_train, yhat_train))
    print("r_2 score (Validation Set): ", r2_score(y_test, yhat))
    
    rse = [rmse(yhat_train,y_train), rmse(yhat, y_test),r2_score(y_train, yhat_train),r2_score(y_test, yhat)]
    if hasattr(clf, 'oob_score_'): rse.append(clf.oob_score_)
      
    print("RMSE and OOB: ",rse)
    rms = sqrt(mean_squared_error(y_test, yhat))
    print("rmse: ",rms)

    f_im = clf.feature_importances_.round(3)
    ser_rank = pd.Series(
        f_im, index=X_train.columns).sort_values(ascending=False)
    print(ser_rank.head(top_features))

    if(show_plot):
        #sns.set(font_scale=3)
        plt.figure(figsize=(30, 40))
        # plt.bar(ser_rank.values,ser_rank.indexBuGn_r
        sns.barplot(y=ser_rank.index, x=ser_rank.values, palette="Blues_r")
        plt.xlabel('relative importance')

    return r2_score(y_test, yhat), rms, clf
```

```python
def xg_boost(X, show_plot=True, top_features=5):
    """
    Extreme Grading Boosting Machines.
    Out of the box implementation of xgboost regressor from the library - xgboost.
    Prints the MSE and r2, along with the topNfeatures
    """
    X_train, X_test, y_train, y_test = test_train_split_custom(X)
    xgb_mod = xgb.XGBRegressor()
    xgb_mod.fit(X_train, y_train)
    yhat = xgb_mod.predict(X_test)
    yhat_train = xgb_mod.predict(X_train)
    
    print('Training Set Performance')
    print('r2_score: ', r2_score(y_train, yhat_train))
    print('RMSE: ', np.sqrt(mean_squared_error(y_train, yhat_train)))
    
    print('Validation Set Performance')
    print('r2_score: ', r2_score(y_test, yhat))
    print('RMSE: ', np.sqrt(mean_squared_error(y_test, yhat)))
    
    rms = sqrt(mean_squared_error(y_test, yhat))

    f_im = xgb_mod.feature_importances_.round(3)
    ser_rank = pd.Series(
        f_im, index=X_train.columns).sort_values(ascending=False)
    print(ser_rank.head(top_features))

    if(show_plot):
        #sns.set(font_scale=3)
        plt.figure(figsize=(30, 40))
        # plt.bar(ser_rank.values,ser_rank.indexBuGn_r
        sns.barplot(y=ser_rank.index, x=ser_rank.values, palette="Blues_r")
        plt.xlabel('relative importance')

    return r2_score(y_test, yhat),rms, xgb_mod
```

```python
def svr(X, drop_feature=False, drop: list = None, kernel = 'rbf', penalty_param = 1, eps = 0.1):
    """
    Performs scikit-learn's implementation of Epsilon-Support Vector Regression.
    Internally, splits the data into test and train and fits a linear regression model on the training set.
    Prints MSE and r2 value after evaluating model on the test set and returns the estimator and the predicted values.
    Has same defaults as sklearn.
    """

    X_train, X_test, y_train, y_test = test_train_split_custom(X)

    if(drop_feature):
        X_train = X_train.drop(drop, axis=1)
        X_test = X_test.drop(drop, axis=1)
        
    print(X_train.columns)

    clf = SVR(C=penalty_param, epsilon = eps)
    clf.fit(X_train, y_train)

    yhat = clf.predict(X_test)
    yhat_train = clf.predict(X_train)
    
    print('Training Set Performance')
    print("MSE :", mean_squared_error(y_train, yhat_train))
    print("r2 :", r2_score(y_train, yhat_train))    
    
    print('Validation Set Performance')
    print("MSE :", mean_squared_error(y_test, yhat))
    print("r2 :", r2_score(y_test, yhat))

    return r2_score(y_test, yhat)
```

```python
def set_rf_samples(n):
    """ Changes Scikit learn's random forests to give each tree a random sample of
    n random rows.
    """
    forest._generate_sample_indices = (lambda rs, n_samples:
        forest.check_random_state(rs).randint(0, n_samples, n))
    
def reset_rf_samples():
    """ Undoes the changes produced by set_rf_samples.
    """
    forest._generate_sample_indices = (lambda rs, n_samples:
        forest.check_random_state(rs).randint(0, n_samples, n_samples))
```

```python
def auto_ml( df_XY, target_label, column_descriptions:dict,
                              type_of_estimator='regressor',
                              automl_kwargs:dict={'verbose' : True, 'ml_for_analytics' : True,
                              'perform_feature_selection' : True},
                              Xcolumns:list=None, seed:int=42, test_size:float=(1/3),
                              verbose:bool=False
                            ):
    ''' Run an automl_pipeline using data in dataframe df_XY, for Y data in columns target_label,
        passing auto_ml the parms in dict automl_kwargs, optionally using only columns Xcolumns for the X data,
        with default train/test proportions 3:1 and random choice seed 42,
        printing out verbose run messages or not, returning or not the trained predictor
        and a dataframe with the top N features as determined by the estimator fitting.
    '''
    
    if Xcolumns is not None:
        try:
            df_XY = df_XY[Xcolumns].copy()
        except KeyError as e:
            logging.error('train_test_feature_score(): Xcolumns has invalid column name(s): {}'.format(e))
            raise
            
    
    df_train, df_test, y_train, y_test = train_test_split(df_XY, df_XY[target_label], test_size=test_size, random_state=seed)

    # Invoke the auto_ml predictor. We always ask for verbose, because if not the topN features would not be generated
    ml_predictor = Predictor( type_of_estimator, column_descriptions,
                              verbose==True)  

    ml_predictor.train( df_train, **automl_kwargs )
    #ml_predictor.train( df_train)
    
    # Score the model on test data
    ml_predictor.score(df_test, df_test[target_label], verbose=3)

    # save the model
    file_name = ml_predictor.save()

    
    trained_model = load_ml_model(file_name)

    pred=[]
    predictions = trained_model.predict(df_test)
    pred.append(predictions)

   
    return ml_predictor, pred
```

```python
def autosklearn (df,time=3600,feature_types):
    """
    Run an automl pipeline using autosklearn. Prints out the models in descending order of performance and r-squared
    """
    X_train, X_test, y_train, y_test = test_train_split_custom ( df )
    
    #Initialization
    automl = autosklearn.regression.AutoSklearnRegressor
        (
            time_left_for_this_task=time,
            #per_run_time_limit=30,
            tmp_folder='/tmp/autosklearn_regression_example_tmp',
            output_folder='/tmp/autosklearn_regression_example_out',
            resampling_strategy='cv',
            resampling_strategy_arguments={'folds': 5},
            seed=42,
        )
        
    # fit() method
        automl.fit(X_train, y_train, dataset_name='boston',
                   feat_type=feature_types)
        
    # Show models that were fit 
        print(automl.show_models())
        
    #predictions
        predictions = automl.predict(X_test)
        
    #r2_squared
        print("R2 score:", sklearn.metrics.r2_score(y_test, predictions))
```

```python
def auto_arima (X):
    """
    Auto ARIMA model for time series forecasting paid parking occupancy. 
    Univariate analysis making the time stamp as the index and having the occupancy as the only attribute. 
    Returns the r2_squared value for the model.
    """
    # make the time stamp as the index and then drop it
    X.index = X['time_stamp']
    X = X.drop('time_stamp',axis=1)

    X_train, X_test, y_train, y_test = test_train_split_custom ( X )
    train = pd.concat([X_train, y_train])
    test = pd.concat([X_test, y_test])
    train = train.drop(['mm_transaction','yyyy_transaction'],axis=1)
    test = test.drop(['mm_transaction','yyyy_transaction'],axis=1)

    model = auto_arima(train)
#     , start_p=1, start_q=1, start_P=1, start_Q=1,
#                   max_p=5, max_q=5, max_P=5, max_Q=5, seasonal=True,
#                   stepwise=True, suppress_warnings=True, D=10, max_D=10,
#                   error_action='ignore')

    # Create predictions for the future, evaluate on test
    preds, conf_int = model.predict(n_periods=test.shape[0], return_conf_int=True)

    # Print the error:
    print("Test RMSE: %.3f" % np.sqrt(mean_squared_error(test, preds)))
    return model
```

#### Linear Regression with Forward Selection

```python
def forward_stepwise_reg(X):
    X_train, X_test, y_train, y_test = test_train_split_custom(X)
    # Store sample size for calculating adjusted R2
    test_n = len(X_test)
    features = list(X_train.columns)
    # Create empty list to store selected feature in each step
    modfeatures = []
    # Create empty list to store R2 for each step
    adj_R2 = []
    # Create a model including up to number of features
    for num in range(len(features)):
        # create empty dataframe to store R2 results
        results = pd.DataFrame(columns = ['R2'])
        # for each, loop through each feature and create model
        for i in range(len(features)):
            # skip any variables already in the model
            if features[i] in modfeatures:
                continue
            # run linear regression on training set
            else:
                # copy list of features and append i
                feats = modfeatures.copy()
                feats.append(features[i])
                clf = LinearRegression()
                clf.fit(X_train[feats], y_train)
                # Calculate R2
                yhat_test = clf.predict(X_test[feats])
                score = r2_score(y_test, yhat_test)
                # Store in results
                results.loc[features[i]] = score
        # once all features have models, find best R2
        best = results.loc[results['R2'].idxmax()]
        # Add to model features
        modfeatures.append(best.name)
        # Store R2
        R2 = best['R2']
        k = len(modfeatures)
        adjR2 = 1 - (1 - R2)*((test_n - 1)/(test_n - (k + 1)))
        adj_R2.append(adjR2)
        # Print results
        print(best.name, adjR2)
        # Break if accuracy decreases
        if num > 0:
            if adj_R2[num] - adj_R2[num-1] < 0.0001:
                # Since accuracy decreased, we want the previous model
                # print the list of features excluding the last one
                print('Final model features', modfeatures[:-1])
                # Print accuracy for the previous iteration
                print('Adjusted R2:', adj_R2[num-1])
                break
```
